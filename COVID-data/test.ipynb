```python
import my_tools as mt
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_cases_total = mt.read_csv('cases_total', 'Total cases')
df_cases_weekly = mt.read_csv('cases_weekly', 'Weekly cases')
```

```python
countries = ['China', 'Belgium', 'South_Korea', 'United_States_of_America']

plt.figure(figsize=(17,9))
for country in countries:
    x = df_cases_total.loc[df_cases_weekly.index, country].tolist()
    y = df_cases_weekly[country].tolist()
    
    plt.plot(x, y, label=country) 

plt.xlabel('weekly new cases')
plt.xscale('log')
plt.ylabel('total cases') 
plt.yscale('log')
plt.legend(loc='upper left')
plt.title('show trend')
plt.show()
```
