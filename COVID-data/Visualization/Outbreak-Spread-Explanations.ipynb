### Total cases and deaths

The following are plots on the total amount of cases per country, per day.

```python
mt.plot_spread('total', countries_of_interest)
```

These are the first graphs that need some more consideration. 

We can make an educated guess that the spread of an outbreak will first go exponentially upward, infecting more people the more people are infected, followed by an exponential dampening, through the same thought process. This can be seen in the case of China and South-Korea. The fact that these countries (as of now) have stopped the spread of the outbreak is the reason they were included in this set. So they can be out reference for how we expect other countries to progress. As another reference we included Iran, who has the outbreak under control but has not yet been able to stop it, but more on that later.

Even though the curve might be simular for most countries the intesity of it is not. This is exactly the feature that details how the actual spread of the outbreak progresses, irregardless of the impacted region.

Do remember that through inperfect testing and the nature of crisis, the data shown will also be inperfect. In reality these curves would most likelly be less steep and more flattened out. But we have to work with what we have.

The difficulty comes when we want to visualize, comprehend and compare the intensity of this kind of exponential growth. Since it is hard to compare the progression of all the countries' outbreak like this.<br>
For this reason, as a first step, we will plot the same data again. But this time we put the y-axes in a logarithmic scale. Since this is the natural scale of an exponential function. Secondly, to make it easier to compare the countries, we will move them allong th x-axes so they all start from the moment they have 100 cases or 10 deaths.

Note that by doing this second step this doesn't show the real-time situation but more how the outbreak progressed through each of the countries from the point where it began for each.

```python
mt.plot_spread('total', countries_of_interest, 'log')
```

Immediatly we get our first intuitive estimate of how countries compare in dealing with the outbreak.<br>
We can see the general trend that every country undergoes by looking at China and South Korea again.

To interpret the graph we can look at the relative position of the countries at any point in time. The higher the angle of a country's line with the x-axes, compared to others, the more drastic the spread at that relative time from their startpoint.<br>
The second interesting thing to look at is the how the slope plateaus for China and South_korea. Since the angle of the line shows growth rate, this plateau represents the moments when thes countries stopped the spread and got the outbreak under controle. We can use this to see if other countries start to plateau like that which would also indicate they're overcoming the spread of the outbreak.<br>
Lastly, the part we're most interested in are the moments where the slope gets lower but not fully plateaus. This means that that country was able to contain the spread and control it. Instead of fully stopping it. Considering the latter might not be feasable for certain countries, it might be usefull to find what kind of control mechanisme work well to at least slow down the spread in future countries.

The graph showing the total deaths can be interpreted in the same way. Though considering there is more variability between countries in here, through healthcare systems, preparedness, ext., there is more variability on the slopes.

Very important to remember whenever you look at graphs on a logarithmic scale is that the distance between 10 and 100 is the same as the distance between 100.000 and 1.0000.000. Meaning that, for instance, even though China plateaued, it did this at around a value of 100.000. So any number of new cases that is below 10.000 will be nearly invisable on this scale. It is for this same reason that we chose to start from 10 and 100 deaths and cases since otherwise the bottom of the graphs are very distorted.

All of which to say that graphs on this scale are the only way to compare the exponetial growth between te countries. But are not great to represent actual numbers. Hence both of these sets of graphs will always be provided in the accompanying notebook.

### Weekly cases compared to total

The final graph we will be going over is a visualization provided by [Minute Physics](#https://www.youtube.com/watch?v=54XLXg4fYsc) on youtube which is very compelling.<br>
It takes the weekly increase in cases on the y-axis and the total amount of cases on the x-axes. This will give us a visualization for how intense the spread is relative to the amount of people already infected. 

Before we plot it, do remember that since there is no axis with time, aka dates, time is actually represented by the countries progression through this graph as they have more total cases. Meaning each point on the line of a certain country is the weekly cases when the total amount of cases was where it is on the x-axes.<br>
This means that every country starts at the bottom left, when they reach 100 cases, and will move to the right of the graph.

```python
mt.plot_trends(countries_of_interest, log=False, glob=False)
```

As we can see, this plot is quite hard to read, since the different population densities and sizes the plots for the differtent countries are to diffent to compare. Since the populutions of countries are limited, the exponential growth can only go so far for South Korea compared to China for example, hence the spread for those two was impossible to see on the previous plot. Though we can see how China and South Korea kind of drop and Iran stabalized a for a while to then continue roughly as before.

Though if we remeber, both the values on the x-axis as well as on the y-axis grow exponentially and as we've seen above, exponential functions are more easilly compared when placing them in a logarithmic scale. So let's do that.

```python
mt.plot_trends(countries_of_interest, log=True)
```

Now the power and elegance of this plot comes to the surface.

Here it is very visable that the level of exponential growth of the outbreak is nearly the same for all the countries. This means that all the countries that follow the same spread of the outbreak will follow the same straight line. This graph shows that the actual spread is roughly the same in most countries in this list. Though again through differing features of each, such as the country size, population density and location of outbreak, this spread is expressed differently in the different countries.

The reason why this graph is compelling is that any country that has stopped their infection spread, or get's it under control, will simply fall off the main line, like China and South-Korea have done. Where China had a hard halt on the outbreak, South-Korea had a periode of relative stabile infection rates before it dropped.<br>
While on the other hand Iran was able to get it under control for a period as well only for the outbreak to potentially grow faster then before.

Though this plot is very handy to see the global trends from the start of their outbreak, though this is all it can do. It cannot be used for any comparissons, since it only looks at the the spread compared to total cases. This also means that in the case China has stopped the first outbreak, but where the infection has recently started again, it will not follow the main line anymore. China will start its new trend if this continues which means that, at a later time, we can compare the main trend of infection for the initial infection to that of resurgence of the infection.

In conclussion, this graph can be used to see how countries experience the initial spread of the outbreak. If they fall under the line, they're containing the virus well. If they're above the line, this means that the infection is spreading stronger than in other countries.

Though over time more tools will be implimented to analyse this graph and hopefully get usable metrics and information out of it.
