## Active countries

This worksheet is part of a project to analyse and visualise data on the COVID-19 outbreak and will be run once a day.<br>
It shows the data from a list of countries made by taking the intersection of the 10 countries with the most new cases and the 10 countries with the most new deaths. Both over the last week.

```python
import my_tools as mt

amount_days = 6 * 7
countries = mt.find_crit('active')
```

### Table

This table holds the latest information on each of the countries.

- Incidence = New cases / Population
- Prevalence = Total cases / Population

```python
mt.show_table(countries)
```

### Amount of cases, deaths

These plots show the amount of cases and deaths over time.

```python
mt.plot_spread('cases', '3days', countries, days=amount_days)
mt.plot_spread('cases', 'total', countries, days=amount_days)
```

Since all of these trends are expronential, it is interesting to look at the on a logarithmic scale.This is mainly used to compare their exponential trends, represented by the angle of the line towards the x-axis. For this reason we will translate them over the x-axis in a way that each country starts from the point they had their first 100 cases, or 10 deaths. This makes them line up neatly to compare their angles.

```python
mt.plot_spread('cases', '3days', countries, scale='log')
mt.plot_spread('cases', 'total', countries, scale='log')
```

### Incidence and Prevalence

Since incidence and prevalence are relative to the population size, we can add the global information as a reference, mainly in the plot that are on the logarithmic scale, without messing up the scale of our plots.

```python
countries.append('Global')

mt.plot_spread('incidence', '3days', countries, days=amount_days)
mt.plot_spread('incidence', '3days', countries, scale='log')
```

### Trends

Show the growth rate compared to the total amount of cases, both relative to the population size.<br>
