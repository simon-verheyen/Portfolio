# Showcase of all plots

This worksheet holds, for a certain set of countries, nearly all plots that can be generated with the data- and toolsets. 

```python
import my_tools as mt

country_list = ['China', 'South Korea', 'United States of America', 'Belgium', 'United Kingdom', 'Italy', 'Spain']
```

3days: The average value over 3 days time<br>
Weekly: The average value over a week


- Prevalence = Total cases / Population
- Incidence =  New cases / Popultaion
- Mortality = Total cases / Total deaths

```python
mt.plot_spread('cases', 'daily', countries=country_list)
mt.plot_spread('cases', '3days', countries=country_list)
mt.plot_spread('cases', 'weekly', countries=country_list)
```

```python
mt.plot_spread('cases', 'daily', country_list, scale='log')
mt.plot_spread('cases', '3days', country_list, scale='log')
mt.plot_spread('cases', 'weekly', country_list, scale='log')
```

```python
mt.plot_spread('cases', 'total', countries=country_list)
mt.plot_spread('cases', 'total', country_list, scale='log')
```

```python
mt.plot_spread('incidence', 'daily', countries=country_list)
mt.plot_spread('incidence', '3days', countries=country_list)
mt.plot_spread('incidence', 'weekly', countries=country_list)
```

```python
mt.plot_spread('incidence', 'daily', country_list, scale='log')
mt.plot_spread('incidence', '3days', country_list, scale='log')
mt.plot_spread('incidence', 'weekly', country_list, scale='log')
```

```python
mt.plot_trends(country_list, subj='spread', per='3days')
mt.plot_trends(country_list, subj='impact', per='3days')
```
