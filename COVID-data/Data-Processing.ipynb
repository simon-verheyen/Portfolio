```python
import pandas as pd
import datetime

df_cases_new = pd.read_csv('data/cases_new.csv')
df_cases_new['Date'] = df_cases_new['Date'].astype('datetime64[ns]')
df_cases_new = df_cases_new.set_index('Date')

df_deaths_new = pd.read_csv('data/deaths_new.csv')
df_deaths_new['Date'] = df_deaths_new['Date'].astype('datetime64[ns]') 
df_deaths_new = df_deaths_new.set_index('Date')

df_populations = pd.read_csv('data/populations_2018.csv').set_index('Countries')

dates = df_cases_new.index
countries = df_cases_new.columns

today = datetime.date.today() - datetime.timedelta(days=1)
```

# Create total

```python
df_cases_total = df_cases_new.copy()
df_deaths_total = df_deaths_new.copy()

for country in countries:
    cases = 0
    deaths = 0
    
    for date in dates:
        cases += df_cases_new.at[date, country]
        deaths += df_deaths_new.at[date, country]
        
        df_cases_total.at[date, country] = cases
        df_deaths_total.at[date, country] = deaths
        
display(df_cases_total)
display(df_deaths_total)
```

# Create weekly

```python
dates_weekly = []

start_date = today
end_date = start_date - datetime.timedelta(days=7)

while end_date > df_cases_new.index[0]:
    dates_weekly.append(start_date)
    
    start_date = end_date
    end_date = start_date - datetime.timedelta(days=7)

dates_weekly.reverse()

struct_cases_weekly = [('Date', dates_weekly)]
struct_deaths_weekly = [('Date', dates_weekly)]

for country in countries:
    cases_weekly = []
    deaths_weekly = []
    
    for date in dates_weekly:
        end_date = date - datetime.timedelta(days=7)
        
        cases_weekly.append(df_cases_new.loc[end_date:date, country].sum())
        deaths_weekly.append(df_deaths_new.loc[end_date:date, country].sum())
    
    struct_cases_weekly.append((country, cases_weekly))
    struct_deaths_weekly.append((country, deaths_weekly))

d_cases = {title: content for (title, content) in struct_cases_weekly}
d_deaths = {title: content for (title, content) in struct_deaths_weekly}

df_cases_weekly = pd.DataFrame(d_cases).set_index('Date')
df_deaths_weekly = pd.DataFrame(d_deaths).set_index('Date')

display(df_cases_weekly)
display(df_deaths_weekly)
```

# Create Relative

```python
df_cases_relative = df_cases_new.copy().astype('float')
df_deaths_relative = df_deaths_new.copy().astype('float')

for country in countries:
    ok = (df_populations.at[country, 'Populations'] != 0)
    
    for date in dates:
        if df_cases_total.at[date, country] != 0:
            rel_value = df_deaths_total.at[date, country] / df_cases_total.at[date, country]
            df_deaths_relative.at[date, country] = rel_value
        else:
            df_deaths_relative.at[date, country] = 0

        if ok:
            df_cases_relative.at[date, country] = df_cases_total.at[date, country] / df_populations.at[country, 'Populations']
        else: 
            df_cases_relative.at[date, country] = 0
            
display(df_cases_relative)
display(df_deaths_relative)
```

# Find country thresholds

```python
struct_threshold = [('ind', ['100_cases', '10_deaths'])]
i = 1

for country in countries:
    cases_start = df_cases_total.loc[df_cases_total[country] > 100]
    
    if len(cases_start > 0):
        info = []
        
        cases_start = cases_start.index[0].date()
        deaths_start = df_cases_total.loc[df_deaths_total[country] > 10]
        
        if len(deaths_start) > 0:
            deaths_start = deaths_start.index[0].date()
        else: 
            deaths_start = today
        
        info.append(cases_start)
        info.append(deaths_start)
        
        struct_threshold.append((country, info))
        
d = {title:content for (title, content) in struct_threshold}
df_thresholds = pd.DataFrame(d).set_index('ind')

display(df_thresholds)
```

# Saving

```python
df_cases_total.to_csv('data/cases_total.csv')
df_cases_weekly.to_csv('data/cases_weekly.csv')
df_cases_relative.to_csv('data/cases_relative.csv')

df_deaths_total.to_csv('data/deaths_total.csv')
df_deaths_weekly.to_csv('data/deaths_weekly.csv')
df_deaths_relative.to_csv('data/deaths_relative.csv')

df_thresholds.to_csv('data/threshold_dates.csv')
```
