# Visualization attempts


```python
import pandas as pd
import matplotlib.pyplot as plt
import my_tools as tools

df_cases_new = pd.read_csv('data/cases_new.csv').set_index('Date')
df_cases_new.name= 'New cases'
df_cases_total = pd.read_csv('data/cases_total.csv').set_index('Date')
df_cases_total.name = 'Total cases'

df_deaths_new = pd.read_csv('data/deaths_new.csv').set_index('Date')
df_deaths_new.name = 'New deaths'
df_deaths_total = pd.read_csv('data/deaths_total.csv').set_index('Date')
df_deaths_total.name = 'Total deaths'
```

```python
def plot_df(df, leg=False, size=(15,9), title=''):
    df.plot(figsize=size, legend=leg)
    
    if title != '':
        plt.title(title)
    else:
        plt.title(df.name)
    plt.show()
```

```python
plot_df(df_cases_new)
plot_df(df_deaths_new)

plot_df(df_cases_total)
plot_df(df_deaths_total)
```

```python
df = df_cases_new.tail(14)
df = df.sum(axis=0)

df = df.nlargest(4)

most_cases = df.index.tolist()
df_cases_most_new = df_cases_new[most_cases].tail(30)
df_cases_most_new.name = 'Countries with the most new cases'

plot_df(df_cases_most_new)
df_cases_most_new.tail(10)
```
